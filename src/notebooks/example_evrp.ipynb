```python
from IPython.display import HTML
```

```python
import imageio
import base64

evrp_animation = []
filenames = [
    'images/evrp/output_1.png',
    'images/evrp/output_2.1.png','images/evrp/output_2.2.png',
    'images/evrp/output_3.1.png', 'images/evrp/output_3.2.png',
    'images/evrp/output_4.1.png', 'images/evrp/output_4.2.png',
    'images/evrp/output_5.1.png', 'images/evrp/output_5.2.png',
    'images/evrp/output_6.1.png', 'images/evrp/output_6.2.png',
]
for filename in filenames:
    evrp_animation.append(imageio.imread('../'+ filename))
imageio.mimsave('../images/evrp/evrp.gif', evrp_animation, duration=1500)  # requires imagemagick

b64 = base64.b64encode(open('../images/evrp/evrp.gif','rb').read()).decode('ascii')
display(HTML(f'<img src="data:image/gif;base64,{b64}" />'))
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np


sns.set_theme("notebook")

# Load the cost data
df_cost_train = pd.read_csv('../model/computational_data/avg_train_cost.csv')
df_cost_val = pd.read_csv('../model/computational_data/avg_val_cost.csv')
df_cost_train['DataType'] = 'Training'
df_cost_val['DataType'] = 'Validation'

# Load the reward data
df_reward_train = pd.read_csv('../model/computational_data/avg_train_reward.csv')
df_reward_val = pd.read_csv('../model/computational_data/avg_val_reward.csv')
df_reward_train['DataType'] = 'Training Reward'
df_reward_val['DataType'] = 'Validation Reward'

# Load the length data
df_length_train = pd.read_csv('../model/computational_data/avg_train_length.csv')
df_length_val = pd.read_csv('../model/computational_data/avg_val_length.csv')
df_length_train['DataType'] = 'Training Length'
df_length_val['DataType'] = 'Validation Length'

# Load the penalty data
df_penalty_train = pd.read_csv('../model/computational_data/avg_train_penalty.csv')
df_penalty_val = pd.read_csv('../model/computational_data/avg_val_penalty.csv')
df_penalty_train['DataType'] = 'Training Penalty'
df_penalty_val['DataType'] = 'Validation Penalty'
```

```python

# Concatenate the dataframes
df = pd.concat([df_cost_train, df_cost_val])
df['Epoch'] = df['Step'] // 125

# Remove rows with epoch > 50
df = df[df['Epoch'] <= 50]

# Create the plot
plt.figure(figsize=(8, 6))
colors = plt.cm.coolwarm(np.linspace(0, 1, 20))

sns.lineplot(x='Epoch', y='Value', hue='DataType', palette=[colors[0], colors[18]], data=df)

# Adjust the label size
plt.xlabel('Epoch', fontsize=18)
plt.ylabel('Value', fontsize=18)

# Adjust the ticklabel sizes
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)

# Adjust the title size
plt.title('Training vs Validation Cost \n', size=20)

# Adjust the legend text size
plt.legend(fontsize='16')

plt.tight_layout()
plt.savefig(f"../images/evrp/plot_validation_training_cost.png", dpi=300, bbox_inches='tight')
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np


sns.set_theme("notebook")

# Load the data
df_train = pd.read_csv('evrp_5_logs.csv')
df_val = pd.read_csv('evrp_5_logs_val.csv')
df_train['DataType'] = 'Training'
df_val['DataType'] = 'Validation'

# Concatenate the dataframes
df = pd.concat([df_train, df_val])
df['Epoch'] = df['Step'] // 125

# Create the plot
plt.figure(figsize=(6, 4))
colors = plt.cm.coolwarm(np.linspace(0, 1, 20))

sns.lineplot(x='Epoch', y='Value', hue='DataType', palette=[colors[17], colors[3]], data=df)

plt.title('Training vs Validation Cost \n')
plt.savefig(f"../images/evrp/plot_validation_training_cost.png")
plt.show()
```
